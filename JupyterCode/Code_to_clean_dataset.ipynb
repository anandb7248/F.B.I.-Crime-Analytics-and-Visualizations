```python
import pandas as pd
```

```python
data = pd.read_excel("table5.xls",skiprows=3)

data.columns
```

```python
data
subset = data.drop('State',axis=1)
subset
```

```python
subset = subset[subset["Area"] == "State Total"].reset_index()
subset
```

```python
states = pd.DataFrame(data['State'].dropna())
states = states['State'].str.replace('\d+', '')
states = pd.DataFrame(states)
states = states[states["State"]!="PUERTO RICO"]
states = states[states["State"]!="DISTRICT OF COLUMBIA"]
states = states.reset_index().drop('index',axis=1)
states
```

```python
merged = states.merge(subset,left_index=True,right_index=True)
merged = merged.drop(['index','Area','Unnamed: 2','Rape\n(legacy\ndefinition)3'],axis=1)
merged.columns = merged.columns.str.replace('\d+', '')
merged
```

```python
for i in range(0,50):
    for name in merged.ix[:,"Violent\ncrime":]:
        merged.ix[i,name] = merged.ix[i,name]/100000
```
